```python
import pandas as pd
df = pd.read_csv("All_Journals_CLEANED.csv")
```

```python
import re
import pandas as pd

def clean_author(name):
    if pd.isna(name):
        return None
    
    name = name.strip()
    
    # remove dots
    name = name.replace(".", "")
    
    # remove brackets
    name = re.sub(r"[()]", "", name)
    
    # normalize spaces
    name = re.sub(r"\s+", " ", name)
    
    parts = name.split()
    
    # FIX: convert PK → P K
    if len(parts) > 1:
        first = parts[0]
        if first.isalpha() and first.isupper() and len(first) <= 3:
            parts[0] = " ".join(list(first))
    
    return " ".join(parts)
```

```python
df["Standardized_Author"] = df["Author_Name"].apply(clean_author)
df["Standardized_Author"] = df["Standardized_Author"].str.title()
```
