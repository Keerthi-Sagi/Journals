```python
import pandas as pd
import numpy as np
```

```python
raw_file = "Information_Systems_article_data.csv" 
df = pd.read_csv(raw_file)

print("Original shape:", df.shape)
df.head()
```

```python
# Remove columns with NaN headers
df = df.loc[:, df.columns.notna()]

# Strip column names
df.columns = df.columns.str.strip()

# Replace empty strings with NaN
df.replace(r'^\s*$', np.nan, regex=True, inplace=True)

# Add Journal_Title column if missing
df["Journal_Title"] = "Information Systems"

# Remove duplicates
if "DOI" in df.columns:
    df = df.drop_duplicates(subset=["DOI"], keep="first")
else:
    df = df.drop_duplicates(keep="first")

# Convert NaN → None
df = df.where(pd.notna(df), None)

print("After cleaning:", df.shape)
```

```python
df.to_csv("Information_Systems_basic_cleaned.csv", index=False)
print("Basic cleaned file saved.")
```
